## Aula 2: Análise exploratória dos dados

Para os desafios dessa aula, vamos utilizar a mesma base de dados que estamos utilizando para o desenvolvimento do nosso projeto. Sendo assim, na célula abaixo vamos realizar a importação dessa base de dados e aplicar as transformações já realizadas até o momento da aula 2.

```python
import pandas as pd
```

```python
# importando a base de dados
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep=';')
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
                      'Prédio Inteiro', 'Loja/Salão',
                      'Galpão/Depósito/Armazém',
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']

# removendo imóveis comerciais
df = dados.query('@imoveis_comerciais not in Tipo')

# selecionando apenas imóveis do tipo apartamento
df = df.query('Tipo == "Apartamento"')

# visualizando a base de dados
df.head()
```

### Desafio 1
Calcular a média de quartos por apartamento

```python
df.Quartos.mean()
```

### Desafio 2
Conferir quantos bairros únicos existem na nossa base de dados

```python
len(df.Bairro.unique())
```

### Desafio 3
Analisar quais bairros possuem a média de valor de aluguel mais elevadas

```python
df.groupby('Bairro')[['Valor']].mean().sort_values('Valor',ascending=False)
```

### Desafio 4
Criar um gráfico de barras verticais que apresente os 5 bairros com a média de valor de aluguel mais elevadas

```python
df_bairros_por_valor = df.groupby('Bairro')[['Valor']].mean().sort_values('Valor',ascending=False)
df_bairros_por_valor.head(5)
```

```python
df_bairros_por_valor.head(5).plot(kind='bar', figsize=(10,10))
```

Importe novamente o arquivo [**alunos.csv**](https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv) para realizar os desafios abaixo.

```python
# Importando novamente os dados
import pandas as pd
import numpy as np


url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
df_alunos = pd.read_csv(url)
df_alunos
```

### Desafio 1
Verifique se a base de dados possui dados nulos e caso tenha, realize o tratamento desses dados nulos

```python
df_alunos.isnull().sum()
```

```python
#Nao existem alunos que tem NaN e estao aprovados
selecao_nan_aprovado = (df_alunos['Notas'].isnull()) & (df_alunos['Aprovado'] == True)
df_alunos[selecao_nan_aprovado]
```

```python
#Eu quero tentar colocar um numero aleatorio entre 0 e 5.9 em todos os NaN
'''selecao_nan_reprovado = (df_alunos['Notas'].isnull()) & (df_alunos['Aprovado'] == False)
df_alunos[selecao_nan_reprovado] = df_alunos[selecao_nan_reprovado].fillna(round(np.random.uniform(low=0,high=5.9,size=1)[0],2))
df_alunos'''

#talvez o metodo mais facil sera colocar todos como 0
df_alunos = df_alunos.fillna(0)
df_alunos
```

### Desafio 2
Os alunos "Alice" e "Carlos", não fazem mais parte da turma. Sendo assim, remova-os da base de dados

```python
#se eu colocasse sem o index, na hora de eu fazer drop o codigo vai pensar que eu quero retirar todo o data frame alunos_a_remover, que contem tambem o cabecalho, por isso eu dexo so indicar as linhas com o index
alunos_a_remover = df_alunos.query('Nome == "Alice" |Nome == "Carlos"').index
alunos_a_remover
```

```python
df_alunos.drop(alunos_a_remover, axis=0, inplace=True)
df_alunos
```

### Desafio 3
Aplique um filtro que selecione apenas os alunos que foram aprovados

```python
alunos_aprovados = df_alunos.query('Aprovado == True')
alunos_aprovados
```

```python
selecao_aprovado = df_alunos.Aprovado == True
alunos_aprovados = df_alunos[selecao_aprovado]
alunos_aprovados
```

### Desafio 4
Salve o DataFrame que possui apenas os alunos aprovados em um arquivo csv chamado "alunos_aprovados.csv"

```python
alunos_aprovados.to_csv('Alunos aprovado.csv', index=False)
```

### **Extra**
Ao conferir as notas dos alunos aprovados, notamos que algumas notas estavam incorretas. As alunas que tiraram nota 7.0, na verdade tinham um ponto extra que não foi contabilizado. Sendo assim, substitua as notas 7.0 da base de dados por 8.0. **Dica**: pesquise pelo método **replace**

```python
alunos_aprovados
```

```python
'''alunos_aprovados['Notas'].replace(to_replace=7,value=8,inplace=True)
alunos_aprovados'''

alunos_aprovados = alunos_aprovados.replace(7.0, 8.0)
alunos_aprovados
```
